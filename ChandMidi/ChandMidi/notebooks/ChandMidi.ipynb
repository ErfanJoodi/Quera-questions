## Import libraries

```python
import pandas as pd
import numpy as np
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
```

## Preprocessing and Feature Engineering

```python
# Droping null values
train_df.dropna(inplace= True, axis= 0)

# Define a function to clean the review texts
def preprocess_text(text):
    # Convert to lowercase
    text= text.lower()
    # Remove HTML tags
    text= re.sub(r'<.*?>', '', text)
    # Remove punctuation and numbers
    text= re.sub(r'[^a-z\s]', '', text)
    # Remove extra whitespace
    text= re.sub(r'\s+', ' ', text).strip()
    return text

# Apply the preprocessing function to the 'Review' column of both dataframes
train_df['cleaned_review']= train_df['Review'].apply(preprocess_text)
test_df['cleaned_review']= test_df['Review'].apply(preprocess_text)
```

    شما مجاز هستید از هر مدلی که آموخته‌اید استفاده کنید. به عبارت بهتر، هدف این سوال پیش‌بینی هرچه بهتر متغیر هدف مسئله است!

## Modeling

```python
# Define features (X) and target (y)
X_train= train_df['cleaned_review']
y_train= train_df['Rating']
X_test= test_df['cleaned_review']

# Create a machine learning pipeline
# This pipeline first converts text to TF-IDF vectors and then applies the Ridge regression model.
# TfidfVectorizer parameters:
# - stop_words='english': Removes common English words.
# - ngram_range=(1, 2): Considers both single words and pairs of words (bigrams).
# - max_features=15000: Limits the vocabulary size to the 15,000 most frequent terms.
# Ridge model parameter:
# - alpha= 3.0: Regularization strength. This value was chosen as a reasonable starting point.
model_pipeline= Pipeline([
    ('tfidf', TfidfVectorizer(stop_words= 'english', ngram_range= (1, 2), max_features= 15000)),
    ('ridge', Ridge(alpha= 3.0))
])

# Train the model on the entire training dataset
model_pipeline.fit(X_train, y_train)
```

    پیشنهاد می‌شود با توجه به این معیار، عملکرد مدل خود را بر روی مجموعه داده آموزش یا اعتبارسنجی ارزیابی کنید.

## Evaluation

```python
# To evaluate the model, we can split the training data and check the R2 score.
X_train_split, X_val, y_train_split, y_val= train_test_split(X_train, y_train, test_size= 0.2, random_state= 48)

# Train on the split training data
model_pipeline.fit(X_train_split, y_train_split)

# Make predictions on the validation set
val_predictions= model_pipeline.predict(X_val)

# Calculate the R2 score
r2= r2_score(y_val, val_predictions)
print(f'R2 Score on Validation Set: {r2:.4f}')
```

## Prediction for Test Data and Submission File

```python
# Retrain the model on the full training data for final predictions
model_pipeline.fit(X_train, y_train)

# Predict ratings for the test set
test_predictions= model_pipeline.predict(X_test)

# Post-processing the predictions:
# 1. Round to the nearest integer.
# 2. Clip the values to be within the valid rating range [1, 10].
rounded_predictions= np.round(test_predictions)
clipped_predictions= np.clip(rounded_predictions, 1, 10)

# Create the submission DataFrame as required
submission= pd.DataFrame({'Rating': clipped_predictions.astype(int)})

# Display the first 5 rows of the submission file
print("\nFirst 5 rows of the submission file:")
submission.head()
```
