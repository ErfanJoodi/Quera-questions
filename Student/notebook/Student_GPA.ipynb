## Import libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score
import warnings

# Ignore warnings for cleaner output
warnings.filterwarnings('ignore')
```

## Preprocessing Setup

```python
# Identify feature types
numeric_features= ['StudyTimeWeekly', 'Absences']
categorical_features= ['ParentalSupport']
# Binary features can often be treated as numeric (0/1)
binary_features= ['Tutoring', 'Extracurricular', 'Sports', 'Music', 'Volunteering']

# Create the preprocessing pipelines for numeric and categorical features
numeric_transformer= StandardScaler()
categorical_transformer= OneHotEncoder(handle_unknown= 'ignore')

# Create a column transformer to apply different transformations to different columns
preprocessor= ColumnTransformer(
    transformers= [
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features),
        ('passthrough', 'passthrough', binary_features) # Pass binary features through
    ],
    remainder= 'drop' # Drop any columns not specified
)

# Define target and features
X= df.drop('GPA', axis= 1)
y= df['GPA']

# --- 3. Splitting the Data ---
X_train, X_val, y_train, y_val= train_test_split(X, y, test_size= 0.2, random_state= 48)
```

## Model Definitions

```python
# Define the models and their hyperparameter grids
models= {
    'LinearRegression': {
        'model': LinearRegression(),
        'params': {}
    },
    'Ridge': {
        'model': Ridge(),
        'params': {'regressor__alpha': [0.1, 1.0, 10.0]}
    },
    'Lasso': {
        'model': Lasso(),
        'params': {'regressor__alpha': [0.01, 0.1, 1.0]}
    },
    'RandomForest': {
        'model': RandomForestRegressor(random_state= 48),
        'params': {
            'regressor__n_estimators': [50, 100],
            'regressor__max_depth': [None, 10]
        }
    },
    'GradientBoosting': {
        'model': GradientBoostingRegressor(random_state= 48),
        'params': {
            'regressor__n_estimators': [100, 200],
            'regressor__learning_rate': [0.05, 0.1]
        }
    },
    'SVR': {
        'model': SVR(),
        'params': {
            'regressor__C': [0.1, 1, 10],
            'regressor__kernel': ['linear', 'rbf']
        }
    }
}
```

## Training and Evaluation

```python
results= []
best_models= {}

for model_name, config in models.items():
    print(f"\n--- Training {model_name} ---")
    
    # Create the full pipeline
    pipeline= Pipeline(steps=[('preprocessor', preprocessor),
                               ('regressor', config['model'])])

    # Perform Grid Search using R2 score
    grid_search= GridSearchCV(pipeline, config['params'], cv= 5, scoring= 'r2', n_jobs= -1)
    grid_search.fit(X_train, y_train)

    # Get the best model
    best_model= grid_search.best_estimator_
    best_models[model_name]= best_model

    # Evaluate on the validation set
    y_pred= best_model.predict(X_val)
    mse= mean_squared_error(y_val, y_pred)
    r2= r2_score(y_val, y_pred)

    # Store results
    results.append({
        'Model': model_name,
        'Best Params': grid_search.best_params_,
        'Validation MSE': mse,
        'Validation R2': r2
    })
    print(f"Best parameters for {model_name}: {grid_search.best_params_}")
    print(f"Validation MSE for {model_name}: {mse:.4f}")
    print(f"Validation R2 for {model_name}: {r2:.4f}")
```

## Rank Models

```python
results_df= pd.DataFrame(results).sort_values(by= 'Validation R2', ascending= False).reset_index(drop= True)
results_df
```

```python
# Choose the Best Model
best_model_name= results_df.iloc[0]['Model']
final_model= best_models[best_model_name]
print(f"\n--- Selected Best Model: {best_model_name} ---")

# Training the best model on entire train dataset for better training
final_model.fit(X, y)

# Preprocessing Function for the Test Dataset
def preprocess_test_data(df_test, train_columns):
    """
    Preprocesses the test dataframe to match the training data structure
    and handles missing 'ParentalSupport' values with Moderate.
    """
    # Handle missing ParentalSupport based on GPA
    if df_test['ParentalSupport'].isnull().any():
        print("Found null values in 'ParentalSupport'. Filling them with Moderate.")
       
        # Apply the filling logic
        df_test['ParentalSupport']= df_test['ParentalSupport'].apply(
            lambda x: 'Moderate' if str(x) == 'nan' else x)
        
        print("Null values in 'ParentalSupport' filled.")

    # Ensure test set has the same columns as training set (in the same order)
    # This also removes any columns in df_test that were not in X_train
    test_cols= [col for col in train_columns if col in df_test.columns]
    df_test_aligned= df_test[test_cols]
    
    # Add missing columns if any (e.g., if a dummy variable column is missing)
    for col in train_columns:
        if col not in df_test_aligned.columns:
            df_test_aligned[col]= 0 # Or a suitable default
            
    # Reorder to be certain
    df_test_aligned= df_test_aligned[train_columns]

    return df_test_aligned

# Prepare the test data
X_test_processed= preprocess_test_data(test_df.copy(), X_train.columns)

# Make Final Predictions
print("\n--- Making predictions on the processed test dataset ---")
final_predictions= final_model.predict(X_test_processed)

# Create Submission DataFrame as Requested
print("\n--- Creating submission DataFrame ---")
submission= pd.DataFrame({'GPA': final_predictions})
print("Submission DataFrame created successfully. Here are the first 5 rows:")
submission.head()
```
