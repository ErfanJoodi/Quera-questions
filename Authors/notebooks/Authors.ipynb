## Import libraries

```python
import pandas as pd
import numpy as np
import warnings
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import f1_score
from sklearn.naive_bayes import BernoulliNB
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.pipeline import Pipeline

warnings.filterwarnings('ignore')
```

## Preprocessing and Feature Engineering

```python
# Separate features (X) and the target variable (y)
X= train_df.drop('author', axis= 1)
y= train_df['author']

# The test data consists of the same features
X_test= test_df

# The author names are categorical and need to be converted to numbers for the models.
# We use LabelEncoder for this task.
label_encoder= LabelEncoder()
y_encoded= label_encoder.fit_transform(y)

# We will use a part of our training data to evaluate the models.
# This helps us choose the best model without looking at the final test data.
X_train, X_val, y_train, y_val= train_test_split(X, y_encoded, test_size= 0.2, random_state= 48, stratify= y_encoded)

# Finding the best number of features
pipeline= Pipeline([
    ('select', SelectKBest(score_func= chi2)),
    ('clf', LogisticRegression(max_iter= 1000))
])

param_grid= {
    'select__k': [40, 80, 180, 381]  # experiment with ranges
}

grid= GridSearchCV(pipeline, param_grid, cv= 5, scoring='accuracy', n_jobs= -1)
grid.fit(X_train, y_train)

print("Best k:", grid.best_params_['select__k'])
print("Best CV accuracy:", grid.best_score_)

# pick the top k features based on chi-square
k= grid.best_params_['select__k']
selector= SelectKBest(score_func= chi2, k= k)

# fit only on training data
X_train= selector.fit_transform(X_train, y_train)

# transform validation and test sets
X_val= selector.transform(X_val)
X_test= selector.transform(X_test)
```

## Training and Evaluating Models

```python
# We will store the performance of each model here
model_performance= []

# --- Model 1: Bernoulli Naive Bayes ---
# This model is well-suited for binary feature data like ours.
print("Training Bernoulli Naive Bayes...")
nb_model= BernoulliNB()
nb_model.fit(X_train, y_train)
y_pred_nb= nb_model.predict(X_val)
score_nb= f1_score(y_val, y_pred_nb, average= 'macro')
model_performance.append({'Model': 'Bernoulli Naive Bayes', 'F1 Score': score_nb})
print(f"Bernoulli Naive Bayes F1 Score: {score_nb:.4f}")

# --- Model 2: Logistic Regression ---
# A robust and interpretable linear model.
print("Training Logistic Regression...")
lr_model= LogisticRegression(random_state= 48, solver= 'liblinear')
lr_model.fit(X_train, y_train)
y_pred_lr= lr_model.predict(X_val)
score_lr= f1_score(y_val, y_pred_lr, average= 'macro')
model_performance.append({'Model': 'Logistic Regression', 'F1 Score': score_lr})
print(f"Logistic Regression F1 Score: {score_lr:.4f}")

# --- Model 3: K-Nearest Neighbors ---
# A simple instance-based learning model.
print("Training K-Nearest Neighbors...")
knn_model= KNeighborsClassifier(n_neighbors= 5)
knn_model.fit(X_train, y_train)
y_pred_knn= knn_model.predict(X_val)
score_knn= f1_score(y_val, y_pred_knn, average= 'macro')
model_performance.append({'Model': 'K-Nearest Neighbors', 'F1 Score': score_knn})
print(f"K-Nearest Neighbors F1 Score: {score_knn:.4f}")

# --- Model 4: Random Forest with Hyperparameter Tuning ---
# An ensemble model that often performs very well. We'll find good hyperparameters.
print("Training Random Forest with GridSearchCV...")
rf_params= {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5]
}
rf_grid= GridSearchCV(RandomForestClassifier(random_state= 48), param_grid= rf_params, cv= 3, scoring= 'f1_macro', n_jobs= -1, verbose= 1)
rf_grid.fit(X_train, y_train)
y_pred_rf= rf_grid.predict(X_val)
score_rf= f1_score(y_val, y_pred_rf, average= 'macro')
model_performance.append({'Model': 'Random Forest', 'F1 Score': score_rf})
print(f"Random Forest Best Params: {rf_grid.best_params_}")
print(f"Random Forest F1 Score: {score_rf:.4f}")

# --- Model 5: XGBoost with Hyperparameter Tuning ---
# A powerful gradient boosting model.
print("Training XGBoost with GridSearchCV...")
xgb_params= {
    'n_estimators': [100, 200],
    'learning_rate': [0.1, 0.05],
    'max_depth': [3, 5]
}
# XGBoost requires target labels to start from 0 and be contiguous, which LabelEncoder already does.
xgb_grid= GridSearchCV(XGBClassifier(random_state= 48, use_label_encoder= False, eval_metric= 'mlogloss'), param_grid= xgb_params, cv= 3, scoring= 'f1_macro', n_jobs= -1, verbose= 1)
xgb_grid.fit(X_train, y_train)
y_pred_xgb= xgb_grid.predict(X_val)
score_xgb= f1_score(y_val, y_pred_xgb, average= 'macro')
model_performance.append({'Model': 'XGBoost', 'F1 Score': score_xgb})
print(f"XGBoost Best Params: {xgb_grid.best_params_}")
print(f"XGBoost F1 Score: {score_xgb:.4f}")

# --- Model 6: Support Vector Machine (SVM) ---
# SVM can be effective but slower on large datasets. We use a linear kernel for speed.
print("Training Support Vector Machine (SVC)...")
svc_model= SVC(kernel= 'linear', random_state= 48, probability= True)
svc_model.fit(X_train, y_train)
y_pred_svc= svc_model.predict(X_val)
score_svc= f1_score(y_val, y_pred_svc, average= 'macro')
model_performance.append({'Model': 'SVM (Linear)', 'F1 Score': score_svc})
print(f"SVM (Linear) F1 Score: {score_svc:.4f}")

# --- Model 7: Ensemble with VotingClassifier ---
# Combine the predictions of some of our best models.
print("Training Ensemble VotingClassifier...")
# We use the tuned models for the ensemble
estimators= [
    ('lr', lr_model),
    ('rf', rf_grid.best_estimator_),
    ('xgb', xgb_grid.best_estimator_),
    ('nb', nb_model)
]
voting_model= VotingClassifier(estimators= estimators, voting= 'soft') # Soft voting uses probabilities
voting_model.fit(X_train, y_train)
y_pred_voting= voting_model.predict(X_val)
score_voting= f1_score(y_val, y_pred_voting, average= 'macro')
model_performance.append({'Model': 'Voting Ensemble', 'F1 Score': score_voting})
print(f"Voting Ensemble F1 Score: {score_voting:.4f}")
```

## Final Prediction with the Best Model

```python
# Identify the best model from our results
best_model_name= results_df.iloc[0]['Model']
print(f"The best model is: {best_model_name} with F1 Score: {results_df.iloc[0]['F1 Score']:.4f}")

# Create an instance of the best model based on its name
if best_model_name == 'Random Forest':
    final_model= rf_grid.best_estimator_
elif best_model_name == 'XGBoost':
    final_model= xgb_grid.best_estimator_
elif best_model_name == 'Logistic Regression':
    final_model= LogisticRegression(random_state= 48, solver= 'liblinear')
elif best_model_name == 'Bernoulli Naive Bayes':
    final_model= BernoulliNB()
elif best_model_name == 'SVM (Linear)':
    final_model= SVC(kernel= 'linear', random_state= 48)
elif best_model_name == 'Voting Ensemble':
    final_model= voting_model # Already contains tuned estimators
else: # Default fallback
    final_model= LogisticRegression(random_state= 48, solver= 'liblinear')

# Train the final model on the ENTIRE training dataset for best performance
final_model.fit(X, y_encoded)
print("Best model retrained on the full training dataset.")

# Make predictions on the competition test data
test_predictions_encoded= final_model.predict(X_test)

# Convert the numeric predictions back to the original author names
test_predictions= label_encoder.inverse_transform(test_predictions_encoded)

# Create the submission file in the specified format
submission_df= pd.DataFrame({'author': test_predictions})
submission_df.to_csv('submission.csv', index=False)

print("\nStep 6: Submission file 'submission.csv' created successfully!")
print("Submission file head:")
submission_df.head()
```
