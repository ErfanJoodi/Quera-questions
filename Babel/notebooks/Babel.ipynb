## Preprocessing

```python
# Dropping null values
train_data.dropna(axis= 0, inplace= True)
```

```python
# Combine title and description into a single text feature
train_data['text']= train_data['title'] + ' ' + train_data['description']
test_data['text']= test_data['title'] + ' ' + test_data['description']

# Define features (X) and labels (y)
X= train_data['text']
labels= ['Thriller', 'Classics', 'Romance', 'Mystery', 'Science', 'Literature', 'Fantasy', 'Historical', 'Fiction']
y= train_data[labels]

# Split the training data into a new training set and a validation set (80-20 split)
X_train, X_val, y_train, y_val= train_test_split(X, y, test_size= 0.2, random_state= 48)
```

## Modeling & Evaluation

```python
# Vectorization
# Initialize and fit the vectorizer ONLY on the training data
vectorizer= TfidfVectorizer(stop_words= 'english', min_df= 5, max_df= 0.7)
X_train_tfidf= vectorizer.fit_transform(X_train)
X_val_tfidf= vectorizer.transform(X_val) # Transform validation data with the same vectorizer

# --- Model Training and Evaluation ---
models= {
    "Logistic Regression": LogisticRegression(C= 4, solver= 'liblinear', random_state= 48),
    "Linear SVC": LinearSVC(C= 1, random_state= 48),
    "Random Forest": RandomForestClassifier(n_estimators= 100, random_state= 48, n_jobs= -1)
}

results= {}

for name, model in models.items():
    print(f"Training {name}...")
    # Use OneVsRestClassifier to handle the multi-label nature of the problem
    ovr_classifier= OneVsRestClassifier(model)
    ovr_classifier.fit(X_train_tfidf, y_train)
    
    # Make predictions on the validation set
    predictions= ovr_classifier.predict(X_val_tfidf)
    
    # Calculate the F1 score (micro-average is good for multi-label imbalance)
    score= f1_score(y_val, predictions, average='micro')
    results[name]= score
    print(f"{name} F1 Score (Micro): {score:.4f}\n")

# Create a DataFrame to rank the models
ranking_df= pd.DataFrame(list(results.items()), columns= ['Model', 'F1_Score (Micro)'])
ranking_df= ranking_df.sort_values(by= 'F1_Score (Micro)', ascending= False).reset_index(drop= True)

print("--- Model Performance Ranking on Validation Set ---")
ranking_df
```

## Test

```python
# 1. Re-initialize and fit the vectorizer on the ENTIRE training dataset
full_vectorizer= TfidfVectorizer(stop_words= 'english', min_df= 5, max_df= 0.7)
X_train_full_tfidf= full_vectorizer.fit_transform(train_data['text'])
X_test_tfidf= full_vectorizer.transform(test_data['text'])

# 2. Initialize the best model with the same parameters
best_model= LogisticRegression(C= 4, solver= 'liblinear', random_state= 48)
final_classifier= OneVsRestClassifier(best_model)

# 3. Train the final model on the ENTIRE training dataset
final_classifier.fit(X_train_full_tfidf, train_data[labels])

# 4. Make predictions on the final test data
final_predictions= final_classifier.predict(X_test_tfidf)

# 5. Create the submission DataFrame
submission= pd.DataFrame(final_predictions, columns= labels)
submission.head()
```
